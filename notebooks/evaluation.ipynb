```python
all_stats = pd.read_csv('../social_simulation_seminar/models/test_2.csv')
all_stats
```

```python
all_stats = all_stats[(all_stats['wants-to-park']==True) & (all_stats['price-paid']!= -99)]
all_stats
```

```python
all_stats.describe()
```

```python
strategy = all_stats.groupby('agent-strategy-flag').mean().reset_index()
strategy = strategy.rename(columns = {'agent-strategy-flag':'Strategy', 'utility-value':'Avg Utility'})
strategy
```

```python
x = strategy['Strategy']
y = strategy['Avg Utility']
ax = sns.set_palette(palette='Blues')
ax = sns.barplot(x=x, y=y, data=strategy, palette='Blues')
ax.set_xlabel('Strategy')
ax.set_ylabel('Utility')
ax.set_title('Average Utility per Strategy')
```

```python
income_utility = all_stats.groupby('income-group').mean().reset_index()
income_utility = income_utility.rename(columns = {'income-group':'Income Group', 'utility-value':'Avg Utility'})
income_utility
```

```python
x = income_utility['Income Group']
y = income_utility['Avg Utility']
sns.set_palette(palette='Blues')
ax = sns.barplot(x=x, y=y, data=income_utility, palette = 'Blues')
ax.set_xlabel('Income Group')
ax.set_ylabel('Utility')
ax.set_title('Average Utility per Income Group')
```

```python
ax = sns.relplot()
x = income_utility['Income Group']
y = income_utility['Avg Utility']
sns.set_palette(palette='Blues')
ax = sns.barplot(x=x, y=y, data=income_utility, palette = 'Blues')
ax.set_xlabel('Income Group')
ax.set_ylabel('Utility')
ax.set_title('Average Utility per Income Group')
```
